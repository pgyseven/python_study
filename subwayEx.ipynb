문제) 유임승차 비율이 가장 높은 역은 어디일까?

풀이 : 유임승차 비율 = 유임승차 인원 / 무임승차 인원

```python
import csv

file  = open('subwayFee.csv', 'r', encoding='utf-8')
date = csv.reader(file)
next(date) # header skip

rate = 0
max_rate = 0
max_station = ''

for row in date:
    for i in range(4, 8):
        row[i] = int(row[i]) # 4열 ~ 7열 int 변환

    rate = row[4] / (row[4] + row[6]) 
    if rate > max_rate:
        max_rate = rate
        max_station = row[3]
    if row[3] == '충무로':
        print(row, rate)


    
print('유임승차 비율이 가장 높은 역:', max_station, max_rate)

file.close()
print('🤑')
```

위 풀이에서는 환승역에 대해 유임승차 인원과 무임승차 인원을 합하지 않은 값이다.
다시 환승역에 대해서 합한 인원을 유임승차 비율을 구해보자.

```python
import csv

file = open('subwayFee.csv', 'r', encoding='utf-8')
data = csv.reader(file)
next(data) # header 없애고 출력하기

stationList = {}

for row in data : 
    for i in range(4, 8) : 
        row[i] = int(row[i]) # 4열부터 7열을 int로 변환

    if row[3] not in stationList :  # stationList에 row[3](station name)이 없으면 추가

        # 역 이름을 key값으로 하고, value를 list(4열부터 7열까지의 승차인원)
        stationList[row[3]] = row[4:8] # stationList에 저장하기

    else : # stationList에 row[3](station name)이 있으면 = 환승역
        print(row[3] + "은(는) 환승역이므로 기존의 승차 인원에 새로운 데이터를 누적해야 합니다.")
        tmpList = row[4:8]
        print("새로 읽어온 값 : ", tmpList)

        feeList = stationList[row[3]] # 기존의 역 데이터의 승차인원정보
        print("기존의 값 : ", feeList)

        for i in range(4) :
            feeList[i] += tmpList[i]
        
        print("누적값 : ", feeList)
        stationList[row[3]] = feeList # stationList에 저장하기
        print("------------------------------------------------------------------")


print(stationList)
file.close()
```

```python
import csv

# 1. CSV 파일을 읽고 데이터를 합산합니다.
with open('subwayFee.csv', 'r', encoding='utf-8') as file:
    reader = csv.reader(file)
    next(reader)  # 헤더를 건너뜁니다.

    station_data = {}  # 각 역의 데이터를 저장할 딕셔너리

    for row in reader:
        station_name = row[3]
        if station_name not in station_data:
            station_data[station_name] = row[:4] + [int(x) for x in row[4:8]]
        else:
            # 중복된 역의 데이터를 합산합니다.
            for i in range(4, 8):
                station_data[station_name][i] += int(row[i])

# 2. 유임승차 비율이 가장 높은 역을 계산합니다.
max_rate = 0
max_station = ''

for station, data in station_data.items():
    # 유임승차 비율 계산
    rate = data[4] / (data[4] + data[6])  # 유임승차 / (유임승차 + 무임승차)

    if rate > max_rate:
        max_rate = rate
        max_station = station


print('유임승차 비율이 가장 높은 역:', max_station, max_rate)



```

문제) 모든 역의 유무임 승하차 비율을 파이차트로 나타내보자

```python
import csv
import matplotlib.pyplot as plt

file = open('subwayFee.csv', 'r', encoding='utf-8')
data = csv.reader(file)
next(data) # header 없애고 출력하기

labels = ['유임승차', '유임하차', '무임승차', '무임하차']
colors=['red','blue','green','yellow'] 
plt.rc('font', family='D2Coding')

for row in data : 
    for i in range(4, 8) : 
        row[i] = int(row[i]) # 4열부터 7열을 int로 변환

    plt.title(row[3] + '(' + row[1] + ')의 유뮤암 숭하차 비율') 
    plt.pie(row[4:8], labels=labels, colors=colors, autopct='%.1f%%')
    plt.savefig(row[3]+"_" +  row[1] + ".png")
    plt.show()

        
```
