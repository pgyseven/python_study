pie 차트 (원 그래프): 전체 데이터 중 특정 데이터의 비율(퍼센트)을 구해 시각화 한 것.

```python
import matplotlib.pyplot as plt
label = ['A형', 'B형', 'O형', 'AB형'] # 각 혈액형의 이름
#plt.rcParams['font.family'] = 'D2Coding'
plt.rc('font', family='D2Coding')
plt.title('우리반 혈액형 분포')
colors=['red','blue','green','yellow'] 

# explode 돌출
# autopct: 각 차트 영역에 % 표시
plt.pie([3, 1, 4, 1 ], labels= label, autopct='%.1f%%', explode=(0,0.3,0,0)) 
plt.legend()
plt.show()
```
