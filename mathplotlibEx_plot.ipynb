데이터 시각화 : 데이터를 분석한 후 사용자에게 원하는 정보를 출력할 때 차트나 이미지 형태로 간략화 하여 출력할 필요가 있는데, 이를 데이터 시각화(Data Visualization) 라고 한다.

데이터 시각화를 할 때 가장 많이 사용하는 도구로 matplotlib, seaborn, plotly, ggplot, d3.js, Tableau, Qlik 등이 있다. 

plot차트 : 
연속 데이터 세트가있는 경우 선 차트를 사용합니다. 이는 데이터 포인트 수가 매우 많을 때 (20 개 이상) 일정 기간 동안의 추세 기반 데이터 시각화에 가장 적합합니다.

```python
import matplotlib.pyplot as plt

plt.plot([10, 20, 30, 50, 40]) # 선그래프를 그릴 데이터 제공(x좌표는 : list의 index, y 좌표 : list의 value )
plt.show() # 그래프를 화면에 출력
```

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [10, 100, 1000, 1]) # plot(x좌표값, y좌표값)
plt.show() 
```

```python
import matplotlib.pyplot as plt

plt.title('matplotlib [plot example]')
plt.plot([1, 2, 3, 4], [10, 100, 1000, 1]) # plot(x좌표값, y좌표값)
plt.show() 
```

```python
import matplotlib.pyplot as plt

plt.rcParams['font.family']='D2Coding' # 폰트 설정(한글 사용하려면 반드시 지정해야 한다.)
plt.title('matplotlib [plot 예제]')
plt.plot([1, 2, 3, 4], [10, 100, 1000, 1]) # plot(x좌표값, y좌표값)
plt.show() 
```

```python
import matplotlib.pyplot as plt

plt.rcParams['font.family']='D2Coding' # 폰트 설정(한글 사용하려면 반드시 지정해야 한다.)
plt.title('차트 색상, 라벨 표시')
plt.plot([10, 20, 30, 40], color='red', label='빨간색') 
plt.plot([40, 30, 20, 10], color='green', label='초록색') 
plt.legend() # 라벨 표시
plt.show() 
```

```python
import matplotlib.pyplot as plt

plt.rcParams['font.family']='D2Coding' # 폰트 설정(한글 사용하려면 반드시 지정해야 한다.)
plt.title('차트의 선 스타일 바꾸기')
plt.plot([10, 20, 30, 40], linestyle='dashed', label='dashed')  #dashed = '--'
plt.plot([40, 30, 20, 10], linestyle=':', label='dotted') 
plt.legend() # 라벨 표시
plt.show() 
```

```python
import matplotlib.pyplot as plt

plt.rcParams['font.family']='D2Coding' # 폰트 설정(한글 사용하려면 반드시 지정해야 한다.)
plt.title('차트의 마커 바꾸기')
plt.plot([10, 20, 30, 40], 'r^', label='triangle up')  # triangle up = '^'  marker='.'  이렇게 안쓰면 컬러주면서 쓸수 있다 r^ 이러면 레드 삼각형
plt.plot([40, 30, 20, 10],  '.', label='circle', color='green')  
plt.legend() # 라벨 표시
plt.show() 
```
