```python
hist : 히스토그램 -> 자료의 분포 상태를 직사각형 모양의 막대 그래프로 나타낸것.
```

```python
import matplotlib.pyplot as plt
# 자료의 빈도수 y축에 자동 결정
plt.hist([1, 1, 2, 3, 7, 4, 7, 7, 5, 5, 7, 10, 8, 8, 8,])
plt.show()
```

```python
import matplotlib.pyplot as plt
import random
dice = []
for i in range(100) :
    dice.append(random.randint(1, 6))
plt.hist(dice)
plt.show()
print(dice)
```
